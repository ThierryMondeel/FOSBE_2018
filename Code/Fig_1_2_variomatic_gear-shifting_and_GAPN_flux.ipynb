```python
import pandas as pd
from bokeh.layouts import gridplot
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import Range1d

output_notebook() # Run once to get inline figures

import matplotlib.pyplot as plt

# SET PLOTTING SETTINGS
SMALL_SIZE = 16
MEDIUM_SIZE = 20
BIGGER_SIZE = 24
BIGGEST_SIZE = 28

plt.rc('font', size=MEDIUM_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=MEDIUM_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=BIGGEST_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=MEDIUM_SIZE)    # legend fontsize
plt.rc('figure', titlesize=MEDIUM_SIZE)  # fontsize of the figure title
```

# Figure 1

```python
df = pd.read_excel('../Data/Fig_1.xlsx')
df.head()
```

## Bokeh

```python
x  = df['x']
y1 = df['y1']
y2 = df['y2']
y3 = df['y3']
y4 = df['y4']

TOOLS = "pan,wheel_zoom,box_zoom,reset,save,box_select"

p = figure(tools=TOOLS)

p.circle(x, y1, legend = "Z = 0.5", fill_color=None, line_color="red")
p.line(x, y1,   legend = "Z = 0.5", line_color="red")

p.circle(x, y2, legend = "Z = 2", color="Blueviolet")
p.line(x, y2,   legend = "Z = 2", line_color="Blueviolet")

p.square(x, y3, legend="Z = 4", fill_color=None, line_color="blue")
p.line(x, y3, legend="Z = 4", line_color="blue")

p.square(x, y4, legend="Z = variomatic",color="deepskyblue")
p.line(x, y4, legend="Z = variomatic", line_color="deepskyblue")

p.x_range = Range1d(0, 1)
p.xaxis.axis_label = "∆Ga / -∆Gc"
p.xaxis.axis_label_text_font_size = "16pt"
p.xaxis.major_label_text_font_size = "14pt"

p.y_range = Range1d(0, 2.6)
p.yaxis.axis_label = "Ja / (Lcc ∙ -∆Gc)" # Normalized Anabolic Flux ...
p.yaxis.axis_label_text_font_size = "16pt"
p.yaxis.major_label_text_font_size = "14pt"

p.legend.label_text_font_size = "14pt"

p.plot_height = 400
p.plot_width = int(1.4*p.plot_height)

show(p)
```

## Matplotlib

```python
plt.plot(x,y1, 'k-',label='Z = 0.5', linewidth=2.0) 
plt.plot(x,y2, 'b-', label='Z = 2', linewidth=2.0)
plt.plot(x,y3, 'g-', label='Z = 4', linewidth=2.0) 
plt.plot(x,y4, 'r-', label='Z = variomatic', linewidth=2.0);

plt.axis([0, 1, 0, 2.6])

plt.ylabel(r'$\frac{J_a}{L_{cc} \cdot -\Delta G_c}$')
plt.xlabel(r'$\frac{\Delta G_a}{-\Delta G_c}$')

plt.legend()

fig = plt.gcf()
fig.set_size_inches(8, 5) # A4: 8.3 x 11.7
fig.savefig('../Figures/Figure_1.pdf', bbox_inches='tight')
plt.show()
```

# Figure 2

```python
d = pd.read_excel('../Data/Fig_2.xlsx')

x  = d['Values[ATP/ADP ratio]']
y  = d['Values[GPAN/(GAPDH+GAPN)]']

d.head()
```

## Bokeh

```python
TOOLS = "pan,wheel_zoom,box_zoom,reset,save,box_select"

p = figure(tools=TOOLS)

p.circle(x, y, color="red")

p.x_range = Range1d(0, 130)
p.xaxis.axis_label = "ATP / ADP"
p.xaxis.axis_label_text_font_size = "16pt"
p.xaxis.major_label_text_font_size = "14pt"

p.y_range = Range1d(0, 2.1)
p.yaxis.axis_label = "Fraction of GAPN Flux"
p.yaxis.axis_label_text_font_size = "16pt"
p.yaxis.major_label_text_font_size = "14pt"

p.plot_height = 400
p.plot_width = int(1.4*p.plot_height)

show(p)
```

## Matplotlib

```python
plt.plot(x,y, '.k', linewidth=2.0) 

plt.axis([0, 125, 0, 2.1])

plt.ylabel('Fraction of GAPN flux', fontsize = '24')
plt.xlabel(r'$\frac{ATP}{ADP}$')

fig = plt.gcf()
fig.set_size_inches(8, 6) # A4: 8.3 x 11.7
fig.savefig('../Figures/Figure_2.pdf', bbox_inches='tight')
plt.show()
```
